```python
def week_data(data):
    currency = data[["AUD/US$", "EUR/US$", "NZD/US$", "GBP/US$", "BRL/US$",
    "CAD/US$", "CNY/US$", "HKD/US$", "INR/US$", "KRW/US$", "MXN/US$",
    "ZAR/US$", "SGD/US$", "DKK/US$", "JPY/US$", "MYR/US$", "NOK/US$",
    "SEK/US$", "LKR/US$", "CHF/US$", "TWD/US$", "THB/US$"]]
    
    data["weekday"] = data["Time Serie"].dt.strftime("%A")
    
    result_df = pd.DataFrame(columns=["Currency", "Most_Week"]) #две колонки
    
    for column in currency:
    # Нахождение наиболее часто встречающейся недели (самой высокие показатели)
        max_value_index = data.groupby("weekday")[column].idxmax() #находим индекс с максимальным значенем
        most_common_week = max_value_index.mode().values[0] #нахождение наиболее часто встречающегося дня
        most_common_week = data.loc[most_common_week, "weekday"]
        result_df = result_df.append({"Currency": column, "Most_Week": most_common_week}, ignore_index=True)

    print('День недели, в которой обычно наблюдались самые высокие показатели')
    return result_df

week_data(pd_data)
```

```python
def hist_plot(data, figsize=(30,40)):
    sns.set(style="ticks")
    sns.set_context("talk")
      
    data = data.drop(["Time Serie", "month", "year", "month_year"], axis=1)
    
    
    num_cols = min(4, len(data.columns)) 
    num_rows = (len(data.columns) + num_cols - 1) // num_cols
    
    fig, axs = plt.subplots(num_rows, num_cols, figsize=figsize)
    plt.subplots_adjust(hspace=0.5, wspace=0.4)
    
    for i, col in enumerate(data.columns):
        row_idx = i // num_cols 
        col_idx = i % num_cols 
        ax = axs[row_idx, col_idx]
        sns.histplot(data[col], ax=ax)
        ax.set_title(col)
        ax.set_xlabel("")  
        ax.set_ylabel("")  
        
    for i in range(len(data.columns), num_rows * num_cols):
        row_idx = i // num_cols
        col_idx = i % num_cols
        fig.delaxes(axs[row_idx, col_idx])
    plt.show();

hist_plot(pd_data)
```

```python
def line_plot(data, figsize=(30,40)):
    
    sns.set(style="ticks")
    sns.set_context("talk")
      
    data = data.drop(["month", "year", "month_year"], axis=1)
    
    
    num_cols = min(4, len(data.columns)) 
    num_rows = (len(data.columns) + num_cols - 1) // num_cols
    
    color_palette = sns.color_palette("husl", n_colors=len(data.columns))
    
    fig, axs = plt.subplots(num_rows, num_cols, figsize=figsize)
    plt.subplots_adjust(hspace=0.5, wspace=0.4)
    
    for i, col in enumerate(data.columns):
        row_idx = i // num_cols 
        col_idx = i % num_cols 
        ax = axs[row_idx, col_idx]
        sns.lineplot(x="Time Serie", y=col, color=color_palette[i], data=data, ax=ax)
        ax.set_title(col)
        ax.set_xlabel("")  
        ax.set_ylabel("")  
        
    for i in range(len(data.columns), num_rows * num_cols):
        row_idx = i // num_cols
        col_idx = i % num_cols
        fig.delaxes(axs[row_idx, col_idx])
    plt.show();

line_plot(pd_data)
```

```python
def correlations_high(data):
    corr_matrix= data.corr()
    
    sorted_coef = corr_matrix.unstack().sort_values(ascending=False)
    sorted_coef = sorted_coef[sorted_coef.index.get_level_values(0) != sorted_coef.index.get_level_values(1)] #убирает корреляции между одинаковыми валютными парами
    top_high_correlations = sorted_coef[:7]

    top_columns = top_high_correlations.index.to_frame().iloc[:, 1].unique()#названия валютных пар

    top_correlation_matrix = corr_matrix[top_columns].loc[top_columns]

    plt.figure(figsize=(12, 10))
    sns.heatmap(top_correlation_matrix, annot=True, linewidths=.5, fmt=".1f", cmap="PiYG")
    plt.title("Top 7 High Correlations: Countries/US Dollar", fontsize=16)
    plt.show()

correlations_high(pd_data_country)
```

```python
def plot_box_grid(data,figsize=(20,30)):
    
    
    data_coef = data.drop('month_year', axis=1)
    corr_matrix = data_coef.corr()
    filt = corr_matrix['USD']
            
    num_cols = min(4, len(filt)) 
    num_rows = (len(filt) + num_cols - 1) // num_cols
    
    sns.set(style='whitegrid')
    fig, axs = plt.subplots(num_rows, num_cols, figsize=figsize)
    plt.subplots_adjust(hspace=0.3, wspace=0.4)
    
    for i, col in enumerate(filt.index):
        row_idx = i // num_cols 
        col_idx = i % num_cols 
        ax = axs[row_idx, col_idx]
        ax.boxplot(data_coef[col])
        ax.set_xticklabels([])
        ax.set_xlabel(col, fontsize=15)
        
    for i in range(len(filt.index), num_rows * num_cols):
        row_idx = i // num_cols
        col_idx = i % num_cols
        fig.delaxes(axs[row_idx, col_idx])
    plt.show();

plot_box_grid(groupby_data)

'''Возмем например евро. Данные имеют достаточную вариабельность от 0.76 до 0.9, а также выбросы выше 1.1,
что может указывать на некоторую волатильность в курсе евро к доллару. Медиана (0.82) находится ближе к нижнему квартилю,
что может указывать на то, что большая часть данных находится ближе к более низким значениям курса.
Присутствие выбросов может свидетельствовать о редких событиях или аномалиях, 
которые могли повлиять на курс валюты в отдельные периоды времени.

Следующий пример с японской иеной. Разница между верхним и нижним квартилями (IQR) составляет 18, 
и диапазон (разница между верхней и нижней границей) составляет 50. 
Исходя из этих данных, можно заключить, что разброс курса японской иены умеренный.
Это может свидетельствовать о более стабильном движении курса.
Медиана (110) находится ближе к верхнему квартилю, что может указывать на то, 
что большая часть данных находится ближе к более высоким значениям курса.
Наличие медианы и границ диапазона (80-130) подчеркивает отсутствие выбросов, что может говорить о том, 
что данные достаточно сгруппированы вокруг основного тренда и нет значительных аномалий.
'''
```
